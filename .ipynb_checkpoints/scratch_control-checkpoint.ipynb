```python
import matplotlib.pyplot as plt 
import numpy as np 
import os
import cv2
from PIL import Image
from time import sleep 

%matplotlib inline
```

```python
# create passive dynamics
# generate pixel values for an image with a wave packet of red light 
# make each time step an image, combine images into a movie 

# compute the wave packet solution for each timestep
def f(x,t,A,sig,spt):
    return A*np.exp(-(1./sig)*(x-spt*t)**2)

A = 1 # amplitude / intensity 
sig = 5 # width (pixels)
spt = 100./3 # pixel-steps per second

x = np.arange(100) # one value per pixel
timesteps = np.linspace(0,5,40)

for step in timesteps:
    plt.plot(x,f(x,step,A,sig,spt))
plt.show()
```

```python
# make image arrays, put them into a list 

# number of frames over three seconds
fps = 24 # frame rate, frames per second
time = 3 # seconds
timesteps = np.linspace(0,3,fps*time)
num_pixels = 1000
A = 255 # amplitude / intensity 
sig = 100 # width (pixels)
spt = num_pixels/3. # pixel-steps per second
y = np.arange(num_pixels)

image_arrays = []
for t in timesteps:
    intensity = f(y,t,A,sig,spt) # np.array((f(y,t,A,sig,spt), np.zeros(num_pixels), np.zeros(num_pixels))).reshape(3,num_pixels).T
    image_array = np.tile(intensity,(num_pixels,1))
    image_arrays.append(image_array)
```

```python
# convert list of float arrays to an mp4 video 

def arrays2video(arrays, video_filename):

    if os.path.exists(video_filename):
        os.remove(video_filename)

    if len(arrays[0].shape) != 2:
        raise ValueError('Image arrays must be 2D')
    else:
        height, width = arrays[0].shape

    video = cv2.VideoWriter(filename=video_filename, fourcc=-1, fps=24, frameSize=(width,height), isColor=False)
    for array in arrays:
        video.write(array.astype('u1'))
    cv2.destroyAllWindows()
    video.release()
```

# 1D bang bang controller 

given an input array, output the bang-bang control signal of the same dimension

```python
''' implements bang-bang control given a state and a threshold

    current state: is a numpy array

    threshold: float value above which inhibition turns on 

    returns: the on or off control signal for each state variable 
'''
def BangBang(current_state, threshold):
    
    if type(current_state) != 'numpy.ndarray':
        current_state = np.array(current_state).astype(float)
   
    if type(threshold) != float:
        raise ValueError("threshold must be a float")    
    
    # a boolean array becomes a 0,1 array
    return np.array(current_state >= threshold).astype(float)
```

```python
''' sigmoid function (0,1) 

    takes numpy array argument
    
    domain is roughly -2pi,2pi
'''
def sigmoid(array):
    return np.multiply(1 + np.tanh(np.multiply(array,0.5)),0.5)

''' 1/(1+x) function maps [0,+inf) to [0,1) 

    f(1) = 0.5
'''
def one_over_x(array):
    
    if type(array) != 'numpy.ndarray':
        array = np.array(array).astype(float)
   
    return np.divide(array,np.add(1,array))
```
